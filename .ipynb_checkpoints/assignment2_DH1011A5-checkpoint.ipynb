## About the Dataset
 The dataset here is a segment of the original created by [Luan Moura, Emanuel Fontelles, Vinicius Sampaio, and Mardônio França](https://data.mendeley.com/datasets/3t9vbwxgr5/3). The data contains all songs released in Spotify from 2011 to 2019 distributed by 7 musical genres. The attributes are explained as follows:
 - `acousticness`: Presence of acoustic instruments
 - `danceability`: How suitable a track is for dancing based on a combination of musical elements
 - `artist_name`: The name of the artist or the band. 
 - `track_name`: The title of the song. 
 - `released_date`: Year in which the track was released.
 - `genre`: Musical genre. 
 - `sadness`: How sad sounds the track based on its valence number. 
 - `topic`: Main topic related to the song. 
 - `energy`: Perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy.
 - `track_id`: Unique track identification number.

**Question 1: Importing and Transforming your Data**

Create a code that implement the following instructions: 

- Import the pandas library. 
- Read the data from the file 'music.csv' into a Dataframe object. Use a semicolon (;) as a separator.
- Rename the Dataframe column `release_date` by `release_year`.
- Drop the Dataframe column `energy`.
- Save your new Dataframe in a file `new_music.csv`, using a semicolon as separator (;).
- Display 10 random rows of your DataFrame. 


```python
 
import pandas as pd
# Read the data from the file 'music.csv' into a Dataframe object, semicolon (;) as a separator.
df = pd.read_csv('music.csv', sep=';')

# Rename the Dataframe column release_date by release_year
df = df.rename(columns={'release_date': 'release_year'})

# Drop the Dataframe column energy
df = df.drop('energy', axis=1)

# Save your new Dataframe in a file new_music.csv, using a semicolon as separator (;)
df.to_csv('new_music.csv', sep=';', index=False)

# display 10 random rows
print(df.sample(10))
```

```python
#QUESTION: What are the 10 artists whose songs show top acousticness value in the rock genre?

#CODE - Place your code below to display the section of the Dataframe that includes the result. 
# Filter the data by rock genre
df_rock = df.loc[df["genre"] == "rock"]

# Sort the data by acousticness in descending order
df_rock_sorted_by_acousticness = df_rock.sort_values(by="acousticness", ascending=False)

# Select the top 10 rows
top_ten_acoustic_rock_artists = df_rock_sorted_by_acousticness.iloc[:10, [0]]

# Print the result
top_ten_acoustic_rock_artists

#ANSWER - Answer in a comment below including only the name of the artists. 
# the 10 artists whose songs show top acousticness
 # twenty one pilots
 # sufjan stevens
 # john mayer
 # dave matthews band
 # radiohead
 # alt-j
 # the lumineers
 # the national
 # angel olsen
 # sufjan stevens
```

```python
#QUESTION: How many sad songs have been written in 2019? A song is considered as sad if the sadness score is higher than 0.5 

#CODE - Place your code below. 
# Filter the data by year 2019 and sadness score greater than 0.5
data_sad_2019 = df.loc[(df["release_year"] == 2019) & (df["sadness"] > 0.5)]

# Get the number of rows
num_sad_songs_2019 = data_sad_2019.shape[0]

# Print the result
num_sad_songs_2019

#ANSWER - Answer in a comment below including only the track names.  
# 11 sad songs were written in 2019
```

**Question 3: Plotting your data**

Based on the new dataset obtained in Question 1, answer the following questions in separate cells. Please, follow the instructions in the comments:
- How many songs have been written by genre? Present your result in a tabular format. 
- Plot the number of tracks by genre in a Pie Plot. Don't forget to add a title to your plot. **Bonus:** Can you add the value (number of tracks) to the Pie Plot? 
- Plot the number of 'world/life' tracks over the years. Make sure that the years are plotted in ascending order. Mention in a comment two main observations about the data represented in this plot. Do not forget to add a title to your plot. 

```python

#QUESTION: How many songs have been written by genre? Present your result in a tabular format. 

#CODE - Place your code below. The output should be a table, or a Series object. 

# Group the dataframe by genre and count the number of tracks for each genre
genre_count = df.groupby("genre")["track_id"].count().reset_index()
genre_count

```

```python
#QUESTION: Plot the number of tracks by genre in a Pie Plot. Don't forget to add a title to your plot. 
           # Bonus: Can you add the value (number of tracks) to the Pie Plot?

#CODE - Place your code below. 

# Create a pie plot to show the proportion of tracks by genre
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.pie(genre_count["track_id"], labels=genre_count["genre"], autopct="%1.1f%%")
plt.title("Number of Tracks by Genre in Spotify Dataset")
plt.show()

```

```python
#QUESTION: Plot the number of 'world/life' tracks over the years. Make sure that the years are plotted in ascending order. 
        # Mention in a comment two main observations about the data represented in this plot. 
        # Do not forget to add a title to your plot. 

#CODE: 
# Filter the dataframe by topic
world_life = df[df["topic"] == "world/life"]

# Group the filtered dataframe by released_date and count the number of tracks for each year
world_life_count = world_life.groupby("release_year")["track_id"].count().reset_index()

# Create a line plot to show the trend of 'world/life' tracks over the years
plt.figure(figsize=(8,6))
plt.plot(world_life_count["release_year"], world_life_count["track_id"], marker="o")
plt.title("Number of 'World/Life' Tracks by Year in Spotify Dataset")
plt.xlabel("Release Year")
plt.ylabel("Number of Tracks")
plt.show()

#ANSWER - Include in a comment below two observations. 
 # 1. The genre "world/life" had the most tracks in 2013.
 # 2. The year 2019 had the fewest "world/life" tracks.
```
